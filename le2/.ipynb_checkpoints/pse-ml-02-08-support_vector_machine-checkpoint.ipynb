### Idea: Optimal Margin Classifier

SVMs are based on the idea of finding a hyperplane that best divides a dataset into two classes.

Standard SVM construct a hyperplane, also known as decision boundary, that best divides the input space $X$ into two disjoint regions. (Vladimir Vapnik)

Support vectors are the data points nearest to the hyperplane, the points of a data set that, if removed, 
would alter the position of the dividing hyperplane. Because of this, they can be considered the critical elements of a data set

How do we find the right hyperplane ?

Idea: To separate two sets of data use the widest street approach

**Notation:**

Labels $ y \in \{-1,+1\} $.  Define the decision function g as:

$
g(z)=\left\{\begin{array}{ll} 1, & z >= 0 \\
      -1, & otherwise\end{array}\right. .
$

$ h_{w,b}(x) = g(w^T x + b)$ for  $w \in R^n , b \in R $


$ w^T x + b = \sum_{i=1}^n w_{i}x_{i} + b$

**Definition:**
Functional Margin of the hyperplane defined by (w,b) with respect to $(x^{(i)}, y^{(i)})$ 

$ f^{(i)} = y^{(i)} (w^T x + b) $

If $ y^{(i)} = 1 $ we want $ (w^T x + b) >> 0$.

If $ y^{(i)} = -1 $ we want $ (w^T x + b) << 0$.

Functional Margin with respect to the whole traing set:

$ f = min_{i=1}^{n} f^{(i)}$

A hyperplane is defines as

$$ \mathbf{w}^T * x + b = 0 $$

For all blue points we are looking at the positive hyperplane with 
$$ \mathbf{w}^T * x + b = +1 $$

For all red points we are looking at the negative hyperplane with
$$ \mathbf{w}^T * x + b = -1 $$

The distance between the two hyperplanes should be as big as posible.

$$ \mathbf{w}^T * x_{blue} + b -(\mathbf{w}^T * x_{red} + b) = 2 $$


$$ \mathbf{w}^T * (x_{blue} - x_{red}) = 2 $$

$$ (x_{blue} - x_{red}) = 2 ||{w}^T ||  $$

```python
######## Implementation from scratch

import numpy as np

class Simple_SVM:
    def __init__(self, C=1.0):
        self.C = C # error term
        self.w = 0 # parameter of hyperplane
        self.b = 0 # parameter of hyperplane
        
    # Hinge Loss Function
    def hingeloss(self, w, b, x, y):
        # Regularizer term
        reg = 0.5 * (w * w)

        for i in range(x.shape[0]):
            # Optimization term
            opt_term = y[i] * ((np.dot(w, x[i])) + b)

            # calculating loss
            loss = reg + self.C * max(0, 1-opt_term)
        return loss[0][0]
        
            
    def fit(self, X, Y, batch_size=100, learning_rate=0.001, epochs=1000):
        # The number of features in X
        number_of_features = X.shape[1]

        # The number of Samples in X
        number_of_samples = X.shape[0]

        c = self.C

        # Creating ids from 0 to number_of_samples - 1
        ids = np.arange(number_of_samples)

        # Shuffling the samples randomly
        np.random.shuffle(ids)

        # creating an array of zeros
        w = np.zeros((1, number_of_features))
        b = 0
        losses = []

        # Gradient Descent logic
        for i in range(epochs):
            # Calculating the Hinge Loss
            l = self.hingeloss(w, b, X, Y)

            # Appending all losses 
            losses.append(l)
            
            # Starting from 0 to the number of samples with batch_size as interval
            for batch_initial in range(0, number_of_samples, batch_size):
                gradw = 0
                gradb = 0

                for j in range(batch_initial, batch_initial + batch_size):
                    if j < number_of_samples:
                        x = ids[j]
                        ti = Y[x] * (np.dot(w, X[x].T) + b)

                        if ti > 1:
                            gradw += 0
                            gradb += 0
                        else:
                            # Calculating the gradients

                            #w.r.t w 
                            gradw += c * Y[x] * X[x]
                            # w.r.t b
                            gradb += c * Y[x]

                # Updating weights and bias
                w = w - learning_rate * w + learning_rate * gradw
                b = b + learning_rate * gradb
        
        self.w = w
        self.b = b

        return self.w, self.b, losses 
        
    def predict(self, X):
        
        prediction = np.dot(X, self.w[0]) + self.b # w.x + b
        return np.sign(prediction)
    
```

```python
simple_svm = Simple_SVM()
w, b, losses = simple_svm.fit(X_data,y_data)
print(w, b)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,10))

for i in range(-50,100):
    i = i/10
    X_x = []
    X_y = []
    Y = []
    for j in range(0,110):
        j = j/10
        x = np.array([i,j])
        y = simple_svm.predict(x)
        X_x.append(i)
        X_y.append(j)
        if y > 0:
            Y.append("red")
        else:
            Y.append("blue")
        
    plt.scatter(X_x,X_y, c=Y)

plt.scatter(X_data[:,0], X_data[:,1], c=y_data, cmap=plt.cm.RdYlBu)
plt.show()
```

## Kernel - Radial Basis Function

- see: https://www.pycodemates.com/2022/10/the-rbf-kernel-in-svm-complete-guide.html?utm_content=cmp-true
- see: https://towardsdatascience.com/radial-basis-function-rbf-kernel-the-go-to-kernel-acf0d22c798a

```python
def RBF(X, gamma=None):
    
    # Free parameter gamma
    if gamma == None:
        gamma = 1.0/X.shape[1]
        
    # RBF kernel Equation
    K = np.exp(-gamma * np.sum((X - X[:,np.newaxis])**2, axis = -1))
    
    return K
```

```python
########################## kernel based SVM -  does not work yet ####################
class SVM():
    
    def __init__(self, learning_rate=0.01, num_iterations=100, 
                 C=1.0, kernel='rbf', gamma=0.1):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.C = C
        self.kernel = kernel
        self.gamma = gamma
        
    
    def fit(self, X, y):
        self.X = X
        self.y = y

        n_samples = X.shape[0]

        # Compute the kernel matrix
        kernel_matrix = self._kernel_matrix(X, X)

        # Initialize Lagrange multipliers and bias
        self.alpha = np.zeros(n_samples)
        self.b = 0

        # Gradient descent optimization
        for _ in range(self.num_iterations):
            for i in range(n_samples):
                if self._decision_function(X[i]) * y[i] < 1:
                    self.alpha[i] += self.learning_rate
                self.b += self.learning_rate * y[i]

        # Support vectors are data points with non-zero Lagrange multipliers
        support_vector_indices = self.alpha > 0
        self.alpha = self.alpha[support_vector_indices]
        self.X = X[support_vector_indices]
        self.y = y[support_vector_indices]

    def predict(self, X):
        predictions = []
        for sample in X:
            prediction = self._decision_function(sample)
            predictions.append(np.sign(prediction))
        return np.array(predictions)
    
    def _kernel_matrix(self, X1, X2):
        if self.kernel == 'rbf':
            pairwise_sq_dists = np.square(np.linalg.norm(X1[:, np.newaxis] - X2, axis=-1))
            kernel_matrix = np.exp(-self.gamma * pairwise_sq_dists)
            return kernel_matrix
        else:
            raise ValueError('Invalid kernel. Only "rbf" kernel is supported.')

    def _decision_function(self, X):
        kernel_values = self._kernel_matrix(self.X, X)
        return np.dot(self.alpha * self.y, kernel_values) + self.b
```

Support vector machine produces nonlinear boundaries by constructing a linear boundary in a large, transformed version of the feature space. It is assumed that the class labels are drawn from {−1,1}. As with all linear models, SVMs use separating hyperplanes as the decision boundary between the two classes. In the case of SVMs, the optimization problem of determining these hyperplanes is set up with the notion of margin. Intuitively, a maximum margin hyperplane is one that cleanly separates the two classes, and for which a large region (or margin) exists on each side of the boundary with no training data points in it.

## Implementing soft-margin SVM with Pytorch

$$ min_{w, b}  \frac{1}{2} ||w||^2  + C * \sum_{i=1}^{n} min(0, 1 - y_{i}*(w^Tx_{i}+b)) $$

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
df = pd.DataFrame(data["data"], columns=data["feature_names"])
targets = []
for t in data["target"]:
    if t > 0:
        targets.append(1)
    else:
        targets.append(-1)
df["target"] = targets
```

```python
df.head()
```

```python
from sklearn. model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(data['data'], data['target'], test_size=0.2, random_state=42)
```

```python
import torch
import torch.nn as nn

train_x = torch.tensor(train_x)
train_y = torch.tensor(train_y)
test_x = torch.tensor(test_x)
test_y = torch.tensor(test_y)
```

```python
torch.manual_seed(seed=42)
svm = nn.Linear(train_x.shape[1],1)
```

```python
epochs = 1000
C = 1.0
lr = 0.01
```

```python
optimizer = torch.optim.SGD(svm.parameters(), lr=lr)
```

```python
# Model Training
n_batch = 0

svm.train()
for epoch in range(epochs):
    optimizer.zero_grad()
    output = svm(train_x.float())
    loss = 0.5 * torch.norm(svm.weight.squeeze()) ** 2
    loss += C * torch.clamp(1 - train_y*output, min=0).mean()
    loss.backward()
    optimizer.step ()
    print(epoch, loss)
```
