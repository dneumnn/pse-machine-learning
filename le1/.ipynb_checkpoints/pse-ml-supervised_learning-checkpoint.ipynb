## Excercise 1: 

Load the csv file "../inputs/california_housing.csv" into a list of features X and labels y. 

The california_housing dataset has a sample size of 20640 datapoints with 8 features and one label.

|feature   |description                              |
|----------|-----------------------------------------|
|MedInc    |median income in block group             |
|HouseAge  |median house age in block group          |
|AveRooms  |average number of rooms per household    |
|AveBedrms |average number of bedrooms per household | 
|Population|block group population                   |   
|AveOccup  |average number of household members      |
|Latitude  |block group latitude                     | 
|Longitude |block group longitude                    |

This dataset was obtained from the StatLib repository https://www.dcc.fc.up.pt/~ltorgo/Regression/cal_housing.html

The target variable is the median house value for California districts expressed in hundreds of thousands of dollars ($100,000). The dataset was derived from the 1990 U.S. census, using one row per census block group. A block group is the smallest geographical unit for which the U.S.Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people). A household is a group of people residing within a home. Since the average\nnumber of rooms and bedrooms in this dataset are provided per household, these columns may take surprisingly large values for block groups with few households and many empty houses, such as vacation resorts.


The list of features should be a list of lists. The inner list contains all feature values to the corresponding label. 

Use python module csv: https://docs.python.org/3/library/csv.html#module-csv



```python
import csv
header = []
X = []
with open("../inputs/california_housing.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for i, row in enumerate(csv_reader):
        assert len(row) == 9
        if i == 0:
            header = row
        else:
            X.append(row)
y = [x[-1] for x in X]
print(header)
print(X[0],y[0])
```

```python
import csv
header = {}
X = {}
with open("../inputs/california_housing.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for i, row in enumerate(csv_reader):
        assert len(row) == 9
        if i == 0: #header
            for i, header_item in enumerate(row):
                header[i] = header_item
                X[header_item] = []
        else:
            for i, row_item in enumerate(row):
                X[header[i]].append(row_item)
            

def get_features_label(X: dict, header: list, label:str):
    label = X[label]
    features = [X[key] for key in X if key != label]
    feature_values = [list(vector) for vector in zip(*[feature for feature in features])]
    return feature_values, label
    
header_values = list(header.values())    
X, y = get_features_label(X=X, header=header_values, label=header_values[-1])
print(header_values)
print(X[0],y[0])
```

## Exercise 3:

Visualize Correlation between variables using Pandas.corr() function and matplotlib https://matplotlib.org

```python
corr = df.corr()
corr
```

```python
import matplotlib.pyplot as plt

fig = plt.figure() # add a figure
ax = fig.add_subplot() # add a plot to the figure, returning axis
cax = ax.matshow(corr, cmap='coolwarm', vmin=-1, vmax=1) #Plot the values of a 2D matrix as color-coded image.
fig.colorbar(cax)
ticks = [i for i in range(len(df.columns))]
ax.set_xticks(ticks)       
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
